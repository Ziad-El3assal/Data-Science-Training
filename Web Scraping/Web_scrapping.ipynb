# T1

```python
import requests
from bs4 import BeautifulSoup as bf
import csv
import pandas as pd
import random
import time
```

```python
def req_data(url):
    res=requests.get(url)
    soup=bf(res.text,'html.parser')
    return soup
```

```python
def write_now_t1(data,d):
    with open("T_1.csv",'a',encoding='utf-8') as f:
        writer=csv.writer(f)
        row=[]
        for i in range(1,len(data),2):
            if data[i-1].get_text()=='Dewpoint':
                row.append(data[i].get_text().strip()[6:-1])
            else:
                row.append(data[i].get_text().strip())
        row.append(d)
        writer.writerow(row)
```

```python
with open("T_1.csv",'w',encoding='utf-8') as f:
    writer=csv.writer(f)
    writer.writerow(['Humidity','Wind Speed','Barometer','temp C','Visibility','Last update','Day'])
for days in range(3):
    for hours in range(3):
        soup=req_data()
        dat=soup.find('div',attrs={'id':"current_conditions_detail"})
        data=dat.find_all('td')
        write_now_t1(data,days)
        df=pd.read_csv('T_1.csv')
        print(df)
        time.sleep(random.randint(0,24*60*60-1))
    time.sleep(24*60*60)
        
```

# T2

```python
p=1
with open("T_2.csv",'w',encoding='utf-8') as f:
    writer=csv.writer(f)
    writer.writerow(['Title','Address','Salary /month','Area M2','Rooms','Furinture'])
with open("T_2.csv",'a',encoding='utf-8') as f:
    writer=csv.writer(f)
    while True:
        url=f"https://www.pararius.com/apartments/amsterdam/page-{p}"
        try:
            soup=req_data(url)
        except :
            break
        f=soup.find('ul',attrs={'class':'search-list','data-controller':"search-list"})
        g=f.find_all('section',attrs={'class':"listing-search-item listing-search-item--list listing-search-item--for-rent"})
        row=[]
        for i in g:
            row.append(i.find_all('h2',attrs={'class':"listing-search-item__title"})[0].get_text().strip())
            row.append(i.find_all('div',attrs={'class':'listing-search-item__sub-title'})[0].get_text().strip())
            row.append(i.find('div',attrs={'class':"listing-search-item__price"}).get_text().strip()[0:-9])
            row.append(i.find('li',attrs={'class':"illustrated-features__item--surface-area"}).get_text().strip()[0:-2])
            row.append(i.find('li',attrs={'class':"illustrated-features__item--number-of-rooms"}).get_text().strip()[0:-5])
            if i.find('li',attrs={'class':"illustrated-features__item--interior"}):
                row.append(i.find('li',attrs={'class':"illustrated-features__item--interior"}).get_text().strip())
            writer.writerow(row)
            row=[]
        zx=soup.find('div',attrs={'class':"page__row page__row--pagination"}).find('ul',attrs={'class':"pagination__list"})
        if zx.find('li',attrs={'class':'pagination__item--next'}):
            p+=1
        else :
            break
```

```python
p=1
url=f'http://books.toscrape.com/catalogue/category/books_1/index.html'
soup=req_data(url)
```

```python
lst=soup.find_all('div',attrs={'class':'side_categories'})
```

```python
def fn(x):
    return x.get('href')
def fy(x):
    return x.get_text().strip()
```

```python
lt=lst[0].find_all('ul')
lst_links=lt[1].find_all('a')
st_url='http://books.toscrape.com/catalogue/category'
p=1
en_url=f'/page-{p}.html'
lst_links=list(map(fn,lst_links))
categ=list(map(fy,lt[1].find_all('a')))
final=[]
for i in lst_links:
    if req_data(st_url+i[2:]).find('li',attrs={'class':'next'}) !=None:
        final.append(st_url+i[2:-11]+en_url)
    else:
        final.append(st_url+i[2:])
```

```python
for i in final:
    if 'page' in i:
        p=1
        for j in range(5):
            i=i[:-6]+str(p)+'.html'
            print(i)
            p+=1
```

```python
cur_data=req_data(final[0])
print(final[0])
lst=cur_data.find_all('article',attrs={'class':'product_pod'})
print(lst[0].find('h3').find('a').get('title'))
print(lst[0].find('p').get('class')[1])
print(lst[0].find('p',attrs={'class':'price_color'}).get_text()[2:])
print(lst[0].find('p',attrs={'class':'instock'}).get_text().strip())
```

```python
rate={'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
with open('T_4.csv','w',encoding='utf-8') as f:
    writer=csv.writer(f)
    writer.writerow(['Name','Rate','Price','In Stock','Category'])
    indx=0
    for i in final:
        
        if 'index' in i:
            cur_data=req_data(i)
            lst=cur_data.find_all('article',attrs={'class':'product_pod'})
            
            for cur_book in lst:
                    row=[]
                    row.append(cur_book.find('h3').find('a').get('title'))
                    row.append(rate[cur_book.find('p').get('class')[1]])
                    row.append(cur_book.find('p',attrs={'class':'price_color'}).get_text()[2:])
                    if cur_book.find('p',attrs={'class':'instock'}).get_text().strip()=='In stock':
                        row.append(1)
                    else:
                        row.append(0)
                    row.append(categ[indx])
                    writer.writerow(row)
        else :
            p=1
            while True:
                cur_data=req_data(i)
                lst=cur_data.find_all('article',attrs={'class':'product_pod'})
                for cur_book in lst:
                    row=[]
                    row.append(cur_book.find('h3').find('a').get('title'))
                    row.append(rate[cur_book.find('p').get('class')[1]])
                    row.append(cur_book.find('p',attrs={'class':'price_color'}).get_text()[2:])
                    if cur_book.find('p',attrs={'class':'instock'}).get_text().strip()=='In stock':
                        row.append(1)
                    else:
                        row.append(0)
                    row.append(categ[indx])
                    writer.writerow(row)
                if cur_data.find('li',attrs={'class':'next'}) !=None:
                    p+=1
                    i=i[:-6]+str(p)+'.html'
                else:
                    break
        
                
        indx+=1
            
```
